```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = {
    "Model": ["RNN", "LSTM", "GRU", "BERT", "FinBERT"],
    "Accuracy": [0.78, 0.82, 0.83, 0.84, 0.89],
    "F1 Score": [0.77, 0.81, 0.82, 0.84, 0.89]
}

df = pd.DataFrame(results)

df
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Model",
    y="Accuracy",
    data=df
)

plt.title("Model Accuracy Comparison")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Model",
    y="F1 Score",
    data=df
)

plt.title("Model F1 Score Comparison")

plt.show()
```

# Final Conclusion

Among all models, FinBERT achieved the highest performance
because it is pre-trained specifically on financial text.

Traditional deep learning models like RNN, LSTM, and GRU
performed reasonably well, but transformer-based models
showed superior contextual understanding.

FinBERT is selected as the final production model for
financial sentiment analysis.
